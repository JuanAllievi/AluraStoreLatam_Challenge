### Importación de datos



```python
import pandas as pd
#importo display para poder ver las tiendas
from IPython.display import display

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"
#Cargo los datos
tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

#Visualizo los datos
print("Tienda 1")
display(tienda.head())
print("Tienda 2")
display(tienda2.head())
print("Tienda 3")
display(tienda3.head())
print("Tienda 4")
display(tienda4.head())
```

```python
#Cree las variables y utilice 'set' para luego poder compararlas
columna_tienda1 = set(tienda.columns)
columna_tienda2 = set(tienda2.columns)
columna_tienda3 = set(tienda3.columns)
columna_tienda4 = set(tienda4.columns)

#Creo una condicion para comparar si los nombres son iguales o diferentes
if columna_tienda1 == columna_tienda2 == columna_tienda3 == columna_tienda4:
  print("Las columnas son iguales")
else:
  print("Las columnas son diferentes")
```

```python
#use .dtype con el print para poder ver que tipos de datos tiene cada tienda y el '\n' para dejar un espacio entre cada tienda
print("Tienda 1")
print(tienda.dtypes, "\n")
print("Tienda 2")
print(tienda2.dtypes, "\n")
print("Tienda 3")
print(tienda3.dtypes, "\n")
print("Tienda 4")
print(tienda4.dtypes, "\n")
```

```python
#la estructura esta crea una columna dentro del dataframe.
#La estructura se veereia asi: 'df[columna]=valor'
tienda["Tienda"]= 'tienda1'
tienda2["Tienda"]= 'tienda2'
tienda3["Tienda"]= 'tienda3'
tienda4["Tienda"]= 'tienda4'
```

```python
#unifico ('pd.concat') los valores de los 4 dataframes en un solo dataframe llamado "data", y usamos ignore inde_index para evitar errores.
data=pd.concat([tienda, tienda2, tienda3, tienda4], ignore_index=True)
```

```python
display(data.head())          # Ver las primeras filas
print(data.shape)             # Ver cuántas filas y columnas hay
print(data['Tienda'].value_counts())  # Ver cuántas filas hay por tienda
```

#1. Análisis de facturación



```python
#para ver las columnas que tenemos. Identificamos si es que hay columna con la cantidad de productos vendidos
print(data.columns)
```

```python
#asumimos que cada fila es un producto unico ya que no aparecio una columna llamada "Cantidades vedidas"
data['Total'] = data['Precio']
```

```python
#calculamos la facturacion por tienda y podemos identificar cual es la que genera mas dinero aunque el resultado se muestre en notación científica
facturacion_por_tienda = data.groupby('Tienda')['Total'].sum()
print(facturacion_por_tienda)
```

# 2. Ventas por categoría

```python
#mostramos la cantidad de cada producto vendido por tienda. Podemos ver que en la tienda 1 se vendieron mas muebles que el resto de categorias
categorias_por_tienda = data.groupby(['Tienda', 'Categoría del Producto']).size()
print(categorias_por_tienda)
```

```python
#Convertimos los índices (Tienda y Categoría) en columnas
categorias_top = categorias_por_tienda.reset_index(name='Cantidad')

#Ordenamos de mayor a menor cantidad, dentro de cada tienda
categorias_top = categorias_top.sort_values(['Tienda', 'Cantidad'], ascending=[True, False])

#Tomamos la primera fila de cada tienda (la categoría más vendida)
categorias_mas_vendidas = categorias_top.groupby('Tienda').first().reset_index()
print(categorias_mas_vendidas)
```

# 3. Calificación promedio de la tienda


```python
#Promedio de calificación por tienda
promedio_calificacion = data.groupby('Tienda')['Calificación'].mean().reset_index()
promedio_calificacion.columns = ['Tienda', 'Promedio de Calificación']
display(promedio_calificacion)
```

# 4. Productos más y menos vendidos

```python
productos_vendidos = data["Producto"].value_counts()
```

```python
productos_mas_vendidos = productos_vendidos.idxmax()
cantidad_productos_mas_vendidos = productos_vendidos.max()
```

```python
productos_menos_vendidos = productos_vendidos.idxmin()
cantidad_productos_menos_vendidos = productos_vendidos.min()
```

```python
print(f"El producto más vendido es '{productos_mas_vendidos}' con {cantidad_productos_mas_vendidos} ventas.")
print(f"El producto menos vendido es '{productos_menos_vendidos}' con {cantidad_productos_menos_vendidos} ventas.")
```

# 5. Envío promedio por tienda

```python
promedio_envio = data.groupby('Tienda')['Costo de envío'].mean()
promedio_envio.columns = ['Tienda', 'Promedio de Envío']
display(promedio_envio)
```

# Identificar la tienda de menor rendimiento


```python
#facturacion
facturacion_por_tienda = data.groupby('Tienda')['Total'].sum().reset_index(name='Facturación')
```

```python
#Calificación
calificacion_por_tienda = data.groupby('Tienda')['Calificación'].mean().reset_index(name='Calificación promedio')
```

```python
#Productos vendidos
productos_vendidos = data['Tienda'].value_counts().reset_index()
productos_vendidos.columns = ['Tienda', 'Cantidad vendida']
```

```python
#Cantidad de categorias distintas por tienda
categorias_unicas = data.groupby('Tienda')['Categoría del Producto'].nunique().reset_index(name='Cantidad de Categorías')
```

```python
#Costo de envio
costo_envio_por_tienda = data.groupby('Tienda')['Costo de envío'].mean().reset_index(name='Costo de Envío promedio')
```

```python
#Unifico todos los datos
rendimiento = facturacion_por_tienda.merge(calificacion_por_tienda, on='Tienda') \
                         .merge(productos_vendidos, on='Tienda') \
                         .merge(categorias_unicas, on='Tienda') \
                         .merge(costo_envio_por_tienda, on='Tienda')
```

```python
print(rendimiento)
```

# **GRAFICOS**

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Configuración de estilo
sns.set(style="whitegrid")

# Gráfico de barras - Facturación
plt.figure(figsize=(7,5))
sns.barplot(data=facturacion_por_tienda, x='Tienda', y='Facturación', palette='Blues_d')
plt.title('Facturación Total por Tienda')
plt.ylabel('Facturación ($)')
plt.xlabel('Tienda')
plt.show()
```

```python
plt.figure(figsize=(7,5))
sns.barplot(data=calificacion_por_tienda, x='Tienda', y='Calificación promedio', palette='Greens_d')
plt.title('Calificación Promedio por Tienda')
plt.ylabel('Promedio de Calificación')
plt.xlabel('Tienda')
plt.ylim(3.9, 4.05)  # rango ajustado para ver diferencias
plt.show()
```

```python
plt.figure(figsize=(7,5))
sns.barplot(data=costo_envio_por_tienda, x='Tienda', y='Costo de Envío promedio', palette='Oranges_d')
plt.title('Costo de Envío Promedio por Tienda')
plt.ylabel('Costo de Envío ($)')
plt.xlabel('Tienda')
plt.show()
```
